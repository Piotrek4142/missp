```
from pulp import *
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

fabryka1_slider = FloatSlider(min=0,max=8, value=1, description="dostepnosc fabryki1", style=style)
fabryka2_slider = FloatSlider(min=0,max=24, value=2, description="dostepnosc fabryki2", style=style)
fabryka3_slider = FloatSlider(min=0,max=36, value=5, description="dostepnosc fabryki3", style=style)

def hurtowo(fabryka1 = 8.00,
            fabryka2 = 24.00,
            fabryka3 = 36.00):
    
    problem = LpProblem("maksymalny zysk",LpMaximize)

    x1 = LpVariable("okna aluminiowe P1", 0)
    x2 = LpVariable("okna drewniane P2", 0)

    problem += 3*x1 + 5*x2, "funkcja zysków"
    problem += x1 <= fabryka1, "dostępność fabryki 1"
    problem += 2*x2 <= fabryka2, "dostępność fabryki 2"
    problem += 3*x1 + 2*x2 <= fabryka3, "dostępność fabryki 3"

    problem.solve()
    print("Status:", LpStatus[problem.status])

    for v in problem.variables():
      print(v.name, "=", v.varValue)

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in problem.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))

interact(hurtowo, fabryka1 = fabryka1_slider,
    fabryka2 = fabryka2_slider,
    fabryka3 = fabryka3_slider)
```
